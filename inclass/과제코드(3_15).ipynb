## 온실 가스 배출량과 기온 상승:

가정: 온실 가스 배출량과 기온 상승 사이에는 선형 관계가 있다.
데이터: 여러 해에 걸친 온실 가스 배출량과 기온 상승률 데이터를 수집.
목표: 온실 가스 배출량을 통해 미래의 기온 상승률을 예측.

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression

# 가상의 데이터 생성
np.random.seed(0)  # 일관된 결과를 위해 난수 발생 시드 설정

# 온실 가스 배출량과 기온 상승률 사이에 선형 관계 설정
# y = 0.5 * x + 2 + 노이즈
n_samples = 100
greenhouse_gas_emissions = np.random.uniform(50, 200, n_samples).reshape(-1, 1)  # 50 ~ 200 사이의 값을 가진 100개의 데이터 생성
temperature_increase = 0.5 * greenhouse_gas_emissions + 2 + np.random.normal(0, 5, n_samples).reshape(-1, 1)  # 기온 상승률 생성 (노이즈 추가)

# 시각화
plt.scatter(greenhouse_gas_emissions, temperature_increase)
plt.xlabel("Greenhouse Gas Emissions")
plt.ylabel("Temperature Increase")
plt.title("Relationship between Greenhouse Gas Emissions and Temperature Increase")
plt.grid(True)
plt.show()

# 선형 회귀 모델 학습
model = LinearRegression()
model.fit(greenhouse_gas_emissions, temperature_increase)

# 새로운 값을 예측
new_emission = np.array([[180]])  # 새로운 온실 가스 배출량 값
predicted_temperature_increase = model.predict(new_emission)

print("Predicted Temperature Increase:", predicted_temperature_increase[0][0])
```

```python
## 수업 ppt에 있는 코드

import matplotlib.pyplot as plt
import pandas as pd
from sklearn.linear_model import LinearRegression

data_root = "https://github.com/ageron/data/raw/main/"
lifesat = pd.read_csv(data_root + "lifesat/lifesat.csv")
x = lifesat[["GDP per capita (USD)"]].values
y = lifesat[["Life satisfaction"]].values

lifesat.plot(kind='scatter', grid=True,
             x="GDP per capita (USD)", y="Life satisfaction")
plt.axis([23_500, 62_500, 4, 9])
plt.show()

model = LinearRegression()

model.fit(x, y)

X_new = [[37_655.2]]
print(model.predict(X_new))
```
